```python
# Install necessary libraries
!pip install gradio PyPDF2 requests

# Import required libraries
import gradio as gr
import PyPDF2
import requests

# Set your OpenRouter API key
OPENROUTER_API_KEY = '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'

def analyze_resume(file):
    try:
        # Read the uploaded PDF file
        with open(file.name, 'rb') as f:
            reader = PyPDF2.PdfReader(f)
            text = "".join([page.extract_text() for page in reader.pages])

        if not text.strip():
            return "Error: No readable text found in the PDF."

        # Prepare the request to OpenRouter API
        headers = {
            "Authorization": f"Bearer {OPENROUTER_API_KEY}",
            "Content-Type": "application/json"
        }

        data = {
            "model": "gpt-4",  # Specify the model you want to use
            "messages": [
                {"role": "user", "content": f"Analyze this resume and suggest improvements:\n{text}"}
            ]
        }

        # Send request to OpenRouter API
        response = requests.post("https://api.openrouter.ai/v1/chat/completions", headers=headers, json=data)

        if response.status_code == 200:
            result = response.json()
            return result['choices'][0]['message']['content']
        else:
            return f"Error: {response.status_code} - {response.text}"

    except Exception as e:
        return f"Error: {str(e)}"

# Create Gradio interface
iface = gr.Interface(
    fn=analyze_resume,
    inputs=gr.File(label="Upload your Resume (PDF)"),
    outputs="text",
    title="AI Resume Analyzer",
    description="Upload your resume to get personalized suggestions for improvement.",
)

# Launch Gradio interface
iface.launch(share=True)
```

```python
# Install necessary libraries
!pip install gradio PyPDF2 spacy

# Import required libraries
import gradio as gr
import PyPDF2
import spacy

# Load spaCy's pre-trained NLP model
nlp = spacy.load("en_core_web_sm")

def analyze_resume(file):
    try:
        # Read the uploaded PDF file
        with open(file.name, 'rb') as f:
            reader = PyPDF2.PdfReader(f)
            text = "".join([page.extract_text() for page in reader.pages])

        if not text.strip():
            return "Error: No readable text found in the PDF."

        # Process the resume text using spaCy
        doc = nlp(text)

        # Extract key information
        skills = []
        education = []
        experience = []

        # Identify entities in the text
        for ent in doc.ents:
            if ent.label_ == "ORG":  # Organizations (e.g., companies, universities)
                education.append(ent.text)
            elif ent.label_ == "DATE":  # Dates (e.g., work periods, graduation dates)
                experience.append(ent.text)
            elif ent.label_ == "WORK_OF_ART":  # Skills or achievements (approximation)
                skills.append(ent.text)

        # Generate feedback
        feedback = "### Resume Analysis\n"
        feedback += "**Extracted Skills:**\n" + ", ".join(set(skills)) + "\n\n"
        feedback += "**Education Details:**\n" + ", ".join(set(education)) + "\n\n"
        feedback += "**Work Experience Periods:**\n" + ", ".join(set(experience)) + "\n\n"

        feedback += "**Suggestions:**\n"
        feedback += "- Highlight your skills more prominently.\n"
        feedback += "- Ensure your education and work experience are clearly listed.\n"
        feedback += "- Add relevant keywords based on the job description you're targeting.\n"

        return feedback

    except Exception as e:
        return f"Error: {str(e)}"

# Create Gradio interface
iface = gr.Interface(
    fn=analyze_resume,
    inputs=gr.File(label="Upload your Resume (PDF)"),
    outputs="markdown",
    title="AI Resume Analyzer",
    description="Upload your resume to get insights and suggestions for improvement.",
)

# Launch Gradio interface
iface.launch(share=True)
```

```python
import genai
import os
import re
import io  # For handling file uploads
from google.colab import files  # For file uploads

# Get API key from Colab secrets (recommended) or directly (less secure)
try:
    GENAI_API_KEY = os.environ['GENAI_API_KEY'] # for secrets
except KeyError:
    print("Please set your GENAI_API_KEY as an environment variable or using Colab secrets.")
    GENAI_API_KEY = input("Enter your GenAI API key: ") # less secure

genai.configure(api_key=GENAI_API_KEY)

def analyze_resume(resume_text, job_description):
    """Analyzes a resume against a job description using GenAI."""
    model = genai.model("gemini-pro")
    prompt = f"""Analyze the following resume against the job description... (same prompt)"""
    try:
        response = model.generate_content(prompt)
        return response.text.strip()
    except Exception as e:
        print(f"Error during analysis: {e}")
        return None

def extract_text_from_file(file_content, file_name):
    """Extracts text from uploaded file content."""
    try:
        if file_name.lower().endswith('.txt'):
            return file_content.decode('utf-8')  # Decode bytes to string
        elif file_name.lower().endswith('.pdf'):
            try:
                import PyPDF2
                pdf_file = io.BytesIO(file_content) #create a file-like object
                reader = PyPDF2.PdfReader(pdf_file)
                text = ""
                for page in reader.pages:
                    text += page.extract_text() or ""
                return text
            except ImportError:
                return "PDF support requires PyPDF2. Please install it."
        elif file_name.lower().endswith('.docx'):
            try:
                import docx
                doc = docx.Document(io.BytesIO(file_content))
                text = "\n".join([paragraph.text for paragraph in doc.paragraphs])
                return text
            except ImportError:
                return "DOCX support requires python-docx. Please install it."
        else:
            return "Unsupported file type."
    except Exception as e:
        return f"Error extracting text: {e}"

# Google Colab specific file upload and processing
if __name__ == "__main__":
    print("Upload your resume file:")
    uploaded_resume = files.upload()
    resume_file_name = next(iter(uploaded_resume)) # get file name
    resume_file_content = uploaded_resume[resume_file_name] # get file content

    print("\nUpload your job description file:")
    uploaded_job = files.upload()
    job_file_name = next(iter(uploaded_job))
    job_file_content = uploaded_job[job_file_name]

    resume_text = extract_text_from_file(resume_file_content, resume_file_name)
    job_description_text = extract_text_from_file(job_file_content, job_file_name)

    if resume_text and job_description_text:
        analysis_results = analyze_resume(resume_text, job_description_text)
        if analysis_results:
            print("\nResume Analysis Results:\n", analysis_results)
        else:
            print("\nResume analysis failed.")
    else:
        print("\nCould not load resume or job description.")
```
